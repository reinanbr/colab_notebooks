```python
import sys
from matplotlib.patches import Circle
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
import statistics as stats
plt.style.use('seaborn')
from matplotlib.patches import Circle  # $matplotlib/patches.py

def circle( xy, radius, color="lightsteelblue", facecolor="none", alpha=1, ax=None ,lw=None):
    """ add a circle to ax= or current axes
    """
        # from .../pylab_examples/ellipse_demo.py
    e = Circle( xy=xy, radius=radius,linestyle='--',lw=lw )
    if ax is None:
        ax = plt.gca()  # ax = subplot( 1,1,1 )
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_edgecolor( color )
    e.set_facecolor( facecolor )  # "none" not None
    e.set_alpha( alpha )



plt.figure(figsize=(9, 9))

# Parâmetros do capacitor e grid
V = 9 # Diferença de potencial entre as placas
d =.05  # Distância entre as placas
num_points = 20

# Cria um grid de pontos
x = np.linspace(0.005, d, num_points)
y = np.linspace(-.045,.06 , num_points)
X, Y = np.meshgrid(x, y)

# Calcula o campo elétrico
E =-(V) / (d - x)

# Componentes do campo elétrico em x e y
Ex = np.ones_like(X) * E
Ey = np.zeros_like(Y)

colors = np.log(np.sqrt(Ey**2 + Ex**2))


cm_ = matplotlib.cm.jet


#p1
x1:np.array = np.array([10,30,45,20,20,45,50,20])
y1:np.array = np.array([50,30,20,40,40,15,15,55])

#p3
x3:np.array = np.array([20,70,75,40,30,60,70,30])
y3:np.array = np.array([80,35,20,60,65,30,20,70])

#p2
x2:np.array = np.array([15,45,60,25,10,50,55,25])
y2: np.array = np.array([65,45,5,45,55,20,15,60])

plt.scatter(x1,y1,c='cyan',s=99,label='dpi = 3.06 V')
plt.scatter(x2,y2,c='yellow',s=99,label='dpi = 2.40 V')
plt.scatter(x3,y3,c='red',s=99,label = 'dpi = 1.25 V')



dev_p1 = stats.stdev(x1)
med_p1 = stats.mean(x1)
print('ddp = 3.06 V \n média do x',med_p1,'desvio padrão',dev_p1)



dev_p2 = stats.stdev(y2)
med_p2 = stats.mean(y2)
print('ddp = 2.40 V \n média do x',med_p2,'desvio padrão',dev_p2)



dev_p3 = stats.stdev(y3)
med_p3 = stats.mean(y3)
print('ddp = 1.25 V \n média do x',med_p3,'desvio padrão',dev_p3)






plt.title('Campo Elétrico entre Placas Circulares\ncom Pontos Equipotênciais Medidos em Experimentação\ncom Espaço Amostral de Desvio Padrão',fontweight='bold')

plt.ylabel('Posição x (metros)',fontweight='bold')
plt.xlabel('Posição y (metros)',fontweight='bold')




circle(xy=(-0,0),radius=x1.max(),color='cyan')
circle(xy=(-0,0),radius=x1.max()-3,lw=.9,color='cyan')
circle(xy=(-0,0),radius=x1.max()+3,lw=.9,color='cyan')

circle(xy=(-0,0),radius=y2.max(),color='yellow')
circle(xy=(-0,0),radius=y2.max()-5,lw=.9,color='yellow')
circle(xy=(-0,0),radius=y2.max()+5,lw=.9,color='yellow')

circle(xy=(-0,0),radius=y3.max(),color='red')
circle(xy=(-0,0),radius=y3.max()-5.5,lw=.9,color='red')
circle(xy=(-0,0),radius=y3.max()+5.5,lw=.9,color='red')


plt.grid(linestyle='--',lw=.2,alpha=.9)

plt.legend()
```
