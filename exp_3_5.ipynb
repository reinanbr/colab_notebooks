```python
import numpy as np
import matplotlib.pyplot as plt

def plot_oersted(ax):
    ax.set_title("Experimento de rsted (1820)")
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_aspect('equal')

    # Fio condutor no centro (em z, saindo da tela)
    ax.plot([0, 0], [-5, 5], 'k--', lw=2, label="Fio com corrente")

    # Linhas de campo magntico circulares
    theta = np.linspace(0, 2*np.pi, 200)
    for r in [1, 2, 3, 4]:
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        ax.plot(x, y, 'b')
        ax.arrow(r*np.cos(np.pi/4), r*np.sin(np.pi/4),
                 -0.2*np.sin(np.pi/4), 0.2*np.cos(np.pi/4),
                 head_width=0.2, head_length=0.3, fc='b', ec='b')

    # Bssola representada como vetor
    ax.arrow(3, 0, 1, 0, head_width=0.2, head_length=0.3, fc='r', ec='r', lw=2)
    ax.text(3.5, 0.3, "Bssola", color='r')

def plot_faraday(ax):
    ax.set_title("Experimento de Faraday (1831)")
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_aspect('equal')

    # Espira
    theta = np.linspace(0, 2*np.pi, 200)
    x = 3 * np.cos(theta)
    y = 3 * np.sin(theta)
    ax.plot(x, y, 'orange', lw=2, label="Espira condutora")

    # m se aproximando
    ax.plot([-5, -3], [0, 0], 'k', lw=8)
    ax.text(-5.5, 0.3, "N", color='blue', fontsize=12, fontweight='bold')
    ax.text(-3.2, 0.3, "S", color='red', fontsize=12, fontweight='bold')

    # Corrente induzida (seta circular)
    ax.arrow(0, 3, -0.5, 0, head_width=0.2, head_length=0.3, fc='g', ec='g')
    ax.arrow(-3, 0, 0, -0.5, head_width=0.2, head_length=0.3, fc='g', ec='g')
    ax.arrow(0, -3, 0.5, 0, head_width=0.2, head_length=0.3, fc='g', ec='g')
    ax.arrow(3, 0, 0, 0.5, head_width=0.2, head_length=0.3, fc='g', ec='g')

    ax.text(0, -4.5, "Corrente induzida", color='g')

# Criar figura
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
plot_oersted(ax1)
plot_faraday(ax2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_oersted():
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_title("Experimento de rsted (1820)")
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_aspect('equal')

    # Fio condutor (em z, saindo da tela)
    ax.plot([0, 0], [-5, 5], 'k--', lw=2, label="Fio com corrente")

    # Linhas de campo magntico circulares
    theta = np.linspace(0, 2*np.pi, 200)
    for r in [1, 2, 3, 4]:
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        ax.plot(x, y, 'b')
        # seta indicando sentido do campo (regra da mo direita)
        ax.arrow(r*np.cos(np.pi/4), r*np.sin(np.pi/4),
                 -0.2*np.sin(np.pi/4), 0.2*np.cos(np.pi/4),
                 head_width=0.2, head_length=0.3, fc='b', ec='b')

    # Bssola representada como vetor
    ax.arrow(3, 0, 1, 0, head_width=0.2, head_length=0.3, fc='r', ec='r', lw=2)
    ax.text(3.5, 0.3, "Bssola", color='r')

    ax.legend()
    plt.show()

plot_oersted()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")  # tema preto

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
ax.set_title("Corrente para CIMA (campo anti-horrio)\nExperimento de Oersted (1820)", color="w")

# Fio condutor no centro
ax.plot([0, 0], [-5, 5], color="white", linestyle="--", lw=2)

# Seta indicando sentido da corrente
ax.arrow(0, -4, 0, 2, head_width=0.4, head_length=0.6, fc="white", ec="white", lw=2)
ax.text(0.1, 2.2, "Corrente", color="white")
ax.text(2, 0, "Campo\nMagntico", color="cyan")

# Linhas de campo magntico circulares
theta = np.linspace(0, 2*np.pi, 200)
for r in [1, 2, 3, 4]:
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    ax.plot(x, y, 'cyan')
    ang = np.pi/4
    ax.arrow(r*np.cos(ang), r*np.sin(ang),
             -0.2*np.sin(ang), 0.2*np.cos(ang),
             head_width=0.2, head_length=0.3, fc='cyan', ec='cyan')

# Bssola
ax.arrow(0, 0, -1, 0, head_width=0.2, head_length=0.3, fc='red', ec='red', lw=2)
ax.text(-0.7, 0.3, "Bssola", color='red')

ax.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

plt.style.use("dark_background")

fig, ax = plt.subplots(figsize=(8, 8), facecolor='black')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
ax.set_title("Experimento de ersted: Campo magntico do fio domina o da Terra", color='white', fontsize=14, pad=20)

# Fio condutor vertical no centro
ax.plot([0, 0], [-6, 5], color='white', linestyle='--', lw=3)
ax.text(0.3, 4.5, "Fio com corrente", color='white', fontsize=12)

# Parmetros para as linhas de campo magntico do fio
radii = np.linspace(0.5, 4, 8)
theta = np.linspace(0, 2*np.pi, 300)

# Colormap para gradiente (azul para ciano)
cmap = plt.get_cmap('cool')

for r in radii:
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    # Para criar gradiente ao longo do crculo, plotar segmentos pequenos
    for i in range(len(theta)-1):
        xs = [x[i], x[i+1]]
        ys = [y[i], y[i+1]]
        color = cmap(i / len(theta))
        ax.plot(xs, ys, color=color, lw=2)

# Campo magntico da Terra (seta amarela menor)
ax.arrow(0, 0, 1.5, 0, head_width=0.3, head_length=0.4, fc='red', ec='red', lw=3)
ax.text(.7, 0.4, "Bssola", color='red', fontsize=12)

# Bssola (seta vermelha) prxima ao fio, desviada pelo campo do fio
# A bssola est em (3,0), e seu vetor  tangente ao crculo de raio 3 (campo do fio)
bx, by = 3, 0
angle = np.pi/2  # sentido do campo do fio no ponto (3,0)  tangente ao crculo: (0,1)
# vetor tangente no ponto (3,0)  (0,1)
ax.arrow(bx, by, 1, 1.5, head_width=0.3, head_length=0.4, fc='yellow', ec='yellow', lw=3)
ax.text(bx+0.3, by+1.9, "Campo da Terra", color='yellow', fontsize=12)

# Texto explicativo
text = (
    "O campo magntico gerado pela corrente no fio\n"
    " muito mais intenso localmente que o campo da Terra.\n"
    "Isso faz a bssola se alinhar ao campo do fio,\n"
    "desviando do campo magntico natural da Terra."
)
ax.text(-4.5, -4.5, text, color='white', fontsize=12, bbox=dict(facecolor='black', alpha=0.7, boxstyle='round,pad=0.5'))

# Remover eixos para foco na figura
ax.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")  # tema preto

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect('equal')
ax.set_title("Experimento de Faraday (1831)", color="w")

# Espira condutora
theta = np.linspace(0, 2*np.pi, 200)
x = 3 * np.cos(theta)
y = 3 * np.sin(theta)
ax.plot(x, y, 'orange', lw=2, label="Espira condutora")

# m se aproximando (destacar com cores)
ax.plot([-5, -3], [0, 0], color="white", lw=8, label="m")  # corpo do m em branco
ax.text(-5.3, 0.3, "N", color='cyan', fontsize=12, fontweight='bold')  # polo norte em ciano
ax.text(-3.2, 0.3, "S", color='red', fontsize=12, fontweight='bold')    # polo sul em vermelho

# Corrente induzida (setas circulares ao longo da espira)
corrente_pos = [
    (0, 3, -0.5, 0),
    (-3, 0, 0, -0.5),
    (0, -3, 0.5, 0),
    (3, 0, 0, 0.5)
]
for x0, y0, dx, dy in corrente_pos:
    ax.arrow(x0, y0, dx, dy, head_width=0.2, head_length=0.3, fc='green', ec='green')

ax.text(0, -3.5, "Corrente induzida", color='green')

# Destacar pontos importantes com crculos
ax.plot(0, 0, 'yo', markersize=8, label="Centro da espira")  # centro da espira
ax.plot(-4, 0, 'mo', markersize=8, label="Posio do m")  # posio aproximada do m

# Eixos invisveis
ax.axis("off")
ax.legend(facecolor="black", edgecolor="white", labelcolor="white")
plt.show()
```
