```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Pontos organizados como numpy arrays
points_3v = np.array([
    [0, -27],
    [25, -35],
    [30, -40],
    [41, -50],
    [49, -60],
    [75, -100],
    [60, -80]
])

# Gerando o espelhamento
points_3v_mirror = points_3v * np.array([-1, 1])  # x negativo, y igual
points_3v = np.concatenate((points_3v, points_3v_mirror), axis=0)

# Outros pontos
points_5v = np.array([
    [0, -2],
    [-30, -5],
    [-75, -5],
    [-60, -2],
    [-50, -5],
    [-40, -5],
    [-80, -10]
])

points_5v_mirror = points_5v * np.array([-1, 1])  # x negativo, y igual
points_5v = np.concatenate((points_5v, points_5v_mirror), axis=0)

points_8v = np.array([
    [0, 22],
    [20, 40],
    [30, 31],
    [35, 40],
    [40, 46],
    [40, 90],
    [30, 100]
])

points_8v_mirror = points_8v * np.array([-1, 1])  # x negativo, y igual
points_8v = np.concatenate((points_8v, points_8v_mirror), axis=0)

# Unindo todos os pontos e seus respectivos potenciais
all_points = np.vstack((points_3v, points_5v, points_8v))
potentials = np.array([3]*len(points_3v) + [5]*len(points_5v) + [8]*len(points_8v))

# Criando uma malha (100x100)
grid_x, grid_y = np.mgrid[-100:100:100j, -110:110:100j]

# Interpolação do potencial
grid_potential = griddata(all_points, potentials, (grid_x, grid_y), method='cubic')

# Plotando o mapa de equipotenciais
plt.figure(figsize=(8,6))

# Linhas da origem mais visíveis
plt.axhline(0, color='lightgray', linewidth=1.5)
plt.axvline(0, color='lightgray', linewidth=1.5)

# Contorno de equipotenciais
contour = plt.contourf(grid_x, grid_y, grid_potential, levels=20, cmap='inferno')
plt.colorbar(contour, label='Potencial (V)')

# Adicionando linhas de equipotenciais
plt.contour(grid_x, grid_y, grid_potential,label="linhas de campo", levels=15, colors='white', linewidths=0.7)
# Definindo a grid
plt.xticks(np.arange(-100, 110, 10))
plt.yticks(np.arange(-110, 120, 10))

# Grade a cada 5 unidades
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(5))
plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(5))


plt.grid(True, which='minor', linestyle='--', alpha=0.5)
plt.grid(True, which='major', linestyle='-', alpha=0.8)

# Plotando os pontos originais
plt.scatter(points_3v[:,0], points_3v[:,1], label='3V', color='blue', edgecolor='black', s=100)
plt.scatter(points_5v[:,0], points_5v[:,1], label='5V', color='red', edgecolor='black', s=100)
plt.scatter(points_8v[:,0], points_8v[:,1], label='8V', color='green', edgecolor='black', s=100)

# Rótulos e legendas
plt.legend()
plt.title('Mapa de Equipotenciais de Diferentes Potenciais Elétricos')
plt.xlabel('X (Unidades)')
plt.ylabel('Y (Unidades)')

# Ajuste dos ticks nos eixos para intervalos de 10

# Layout mais compacto
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Dados
vi = np.array([0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10])  # V
amp = np.array([0.3,0.5,0.7,0.9,1.1,1.3,1.5,1.7,1.9,2.1,2.3,2.5,2.7,2.9,3.1,3.4,3.6,3.8,4.0,4.2])  # mA
tens = [0.29,0.52,0.77,1.02,1.26,1.49,1.74,1.94,1.97,2.21,2.46,2.7,2.97,3.18,3.43,3.67,3.96,4.19,4.42,4.65]  # V

# Converter para ampères
amp = amp / 1000

# Modelo: Vi = R * I
def modelo(I, R):
    return R * I

# Ajuste
params, _ = curve_fit(modelo, amp, tens)
resistencia_constante = params[0]

# Plot
I_linspace = np.linspace(min(amp), max(amp), 100)
Vi_fit = modelo(I_linspace, resistencia_constante)

plt.figure(figsize=(10, 10))

# Estilo das linhas
linewidth = 2.5

# Plot dos dados experimentais e do ajuste
plt.plot(tens, amp, 's', color="red", label='Dados experimentais', markersize=7)
plt.plot(Vi_fit, I_linspace, '--', color='black', linewidth=linewidth, label=r'Ajuste Linear ($V = R \cdot I$)' + f"\n$R = {resistencia_constante:.2f} \Omega$")

# Layout científico
plt.xlabel(r'Tensão $V$ (V)', fontsize=24)
plt.ylabel(r'Corrente $I$ (A)', fontsize=24)
plt.title('Tensão e Corrente no Circuito com um Resistor', fontsize=24)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=18)
plt.tick_params(axis='both', labelsize=18)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Dados
vi = np.array([0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10])  # V
amp_mA = np.array([0.3,0.5,0.7,0.9,1.1,1.3,1.5,1.7,1.9,2.1,2.3,2.5,2.7,2.9,3.1,3.4,3.6,3.8,4.0,4.2])  # mA
amp = amp_mA / 1000  # Convertendo para A

# 1. Resistência ponto a ponto
resist_ponto_a_ponto = tens / amp

# 2. Ajuste com curve_fit para resistência constante
def modelo(I, R):
    return R * I

params, _ = curve_fit(modelo, amp, tens)
res_const = params[0]
print(f"Resistência constante ajustada: {res_const:.2f} Ω")

# 3. Plot comparativo
plt.figure(figsize=(10,6))
plt.plot(vi, resist_ponto_a_ponto, 'o-', label='R = Vi / I (ponto a ponto)', color='green')
plt.hlines(res_const, vi[0], vi[-1], colors='orange', linestyles='--', label=f'R ajustada = {res_const:.2f} Ω')

plt.xlabel('Tensão de Entrada $V_i$ (V)')
plt.ylabel('Resistência (Ω)')
plt.title('Comparação entre Resistência Ajustada e Resistência Ponto a Ponto')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Constantes
q = 1.602e-19
k = 1.381e-23
T = 302
I_S = 1e-12

# Função modelo
def corrente_diodo(V, n_):
    return I_S * (np.exp((q * V) / (n_ * k * T)) - 1)

# Dados experimentais
V_dados = np.array([0.5, 1.03, 1.53, 1.71, 1.85, 1.87, 1.89, 1.9, 1.93, 1.94, 1.95, 1.95])
I_dados_mA = np.array([0., 0., 0., 0.1, 0.7, 0.91, 1.3, 1.76, 2.11, 2.53, 2.91, 3.32])
I_dados = I_dados_mA / 1000  # mA → A

# Ajuste
params, _ = curve_fit(corrente_diodo, V_dados, I_dados, p0=[1.0])
n_ajustado = params[0]

# Corrente prevista pelo modelo nos pontos experimentais
I_modelo_mA = corrente_diodo(V_dados, n_ajustado) * 1000  # em mA

# Cálculo dos erros individuais (resíduos absolutos)
erros = np.abs(I_dados_mA - I_modelo_mA)

# Para garantir que pontos exatos (resíduo zero) ainda mostrem uma barra mínima visual,
# definimos um erro mínimo muito pequeno (opcional, mas ajuda na visualização)
erro_minimo = 0.05  # mA
erros = np.where(erros < erro_minimo, erro_minimo, erros)

# Curva suave do modelo
V_suave = np.linspace(min(V_dados), max(V_dados) + 0.02, 300)
I_suave = corrente_diodo(V_suave, n_ajustado) * 1000

# Plotagem
fig, ax = plt.subplots(figsize=(5, 6))

# Dados experimentais com barras de erro individuais
ax.errorbar(V_dados, I_dados_mA, yerr=erros, fmt='o', color='tab:red',
            ecolor='gray', capsize=4, label='Dados experimentais')

# Curva ajustada
ax.plot(V_suave, I_suave, '--', label='Ajuste do modelo', color='black')

# Anotação da equação no gráfico
equacao = (r"$I = I_S \cdot \left(e^{\frac{qV}{n k T}} - 1\right)$" "\n"
           rf"$n$ ajustado ≈ {n_ajustado:.2f}")
ax.annotate(
    equacao,
    xy=(0.6, 1.5),
    xycoords='data',
    fontsize=12,
    bbox=dict(boxstyle="round,pad=0.9", fc="white", ec="gray", lw=1)
)

# Configurações do gráfico
ax.set_xlabel('Tensão (V)', fontsize=12)
ax.set_ylabel('Corrente (mA)', fontsize=12)
ax.set_title('Curva I-V de um Diodo com Barras de Erro Individuais', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()
plt.tight_layout()
plt.show()

# Impressão no terminal
print(f"Fator de idealidade ajustado: n ≈ {n_ajustado:.2f}")
print("Erros individuais (mA):", np.round(erros, 3))
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

# Dados
amp = [0.1, 0.7, 0.91, 1.3, 1.76, 2.11, 2.53, 2.91, 3.32]  # Corrente (mA)
tens = [1.71, 1.85, 1.87, 1.89, 1.9, 1.93, 1.94, 1.95, 1.95]  # Tensão no sistema

# Modelo linear no logaritmo: V = R * ln(I)
def modelo_log(I, R, C):
    return R * np.log(I) + C

# Ajuste de curva
params, _ = curve_fit(modelo_log, amp, tens)
R_ajustado, C_ajustado = params

# Gerando pontos para a linha de ajuste
amp_linspace = np.linspace(min(amp), max(amp), 100)
tens_ajustado = modelo_log(amp_linspace, R_ajustado, C_ajustado)

# Cálculo do R²
residuos = tens - modelo_log(np.array(amp), R_ajustado, C_ajustado)
ss_res = np.sum(residuos**2)
ss_tot = np.sum((tens - np.mean(tens))**2)
r_squared = 1 - (ss_res / ss_tot)

# Criando figura e eixo principal
fig, ax2 = plt.subplots(figsize=(10, 6))

# Eixo Y direito - Tensão no sistema
ax2.set_xlabel("Corrente (mA)", color="tab:gray")
ax2.set_ylabel('Tensão no Sistema (V)', color='tab:blue')
ax2.plot(amp, tens, 'o', color='tab:blue', label='Dados Experimentais')
ax2.plot(amp_linspace, tens_ajustado, 'r-', label=f'Ajuste (R = {R_ajustado:.2f}, C = {C_ajustado:.2f})\n$R^2$ = {r_squared:.4f}')

# Ajustes visuais
ax2.tick_params(axis='y', labelcolor='tab:blue')
ax2.tick_params(axis="x", labelcolor="tab:gray")
plt.title('Corrente e Tensão no Sistema')
ax2.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# --- Dados experimentais ---
I = np.array([0.2, 0.33, 0.86, 1.29, 1.69, 2.05, 2.5, 2.84, 3.31, 3.65, 4.1, 4.5, 4.94]) / 1000  # A
V = np.array([0.0, 0.01, 0.02, 0.02, 0.03, 0.03, 0.04, 0.04, 0.05, 0.06, 0.06, 0.07, 0.08])       # V
R_exp = V / I                                                                                     # Ohm
R0 = R_exp[1]

# --- Modelo físico: R(T) = R0 * [1 + α * (T - T0)] ---
T0 = 28   # temperatura ambiente em K

def R_model(T, R0, alpha):
    return R0 * (1 + alpha * (T - T0))

# --- Ajuste de curva ---
T_guess = T0 + V * I * 1000

# Ajuste de R_model(T) aos dados (T_guess, R_exp)
popt, _ = curve_fit(R_model, T_guess, R_exp, p0=[R0, 0.0045])
R0_fit, alpha_fit = popt

# --- Potência dissipada (Lei de Joule) ---
R_model_fit = R_model(T_guess, R0_fit, alpha_fit)
P = I ** 2 * R_model_fit

# --- Plotagem: Potência vs Temperatura ---
plt.figure(figsize=(10, 7))
plt.plot(T_guess, P, 'o-', color='darkorange', label='Potência dissipada (Lei de Joule)')

# Adicionando a equação no gráfico:
equation = rf"$P = I^2 R(T)$" + "\n" + rf"$R(T) = {R0_fit:.3f} [1 + ({alpha_fit:.2e})(T - {T0})]$"
plt.text(min(T_guess), max(P) * 0.7, equation, fontsize=20,
         bbox=dict(facecolor='white', alpha=0.8))

# Configurações do gráfico
plt.xlabel('Temperatura (ºC)', fontsize=24)
plt.ylabel('Potência Dissipada (W)', fontsize=24)
plt.title('Potência Dissipada vs Temperatura', fontsize=24)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# --- Dados experimentais ---
I = np.array([0.2, 0.33, 0.86, 1.29, 1.69, 2.05, 2.5, 2.84, 3.31, 3.65, 4.1, 4.5, 4.94]) / 1000  # A
V = np.array([0.0, 0.01, 0.02, 0.02, 0.03, 0.03, 0.04, 0.04, 0.05, 0.06, 0.06, 0.07, 0.08])       # V
R_exp = V / I                                                                                     # Ohm
R0 = R_exp[1]

# --- Modelo físico: R(T) = R0 * [1 + α * (T - T0)] ---
T0 = 28   # temperatura ambiente em K

def R_model(T, R0, alpha):
    return R0 * (1 + alpha * (T - T0))

# --- Ajuste de curva ---
T_guess = T0 + V * I * 1000
popt, _ = curve_fit(R_model, T_guess, R_exp, p0=[R0, 0.0045])
R0_fit, alpha_fit = popt

# --- Cálculo da margem de erro ---
R_pred = R_model(T_guess, R0_fit, alpha_fit)
V_fit = I * R_pred
erro = np.abs(V - V_fit)  # Diferença absoluta entre experimental e modelo
err_m = sum(erro)/len(erro)

# --- Plotagem Científica com Margem de Erro ---
plt.figure(figsize=(8, 6))

# Plotando com barras de erro (margem de erro em relação ao modelo)
plt.errorbar(V, I, xerr=erro, yerr=erro / R_pred, fmt='o', color='red',
             ecolor='black', elinewidth=1.5, capsize=4, label='Dados experimentais'+f"\nerro médio = {err_m:.2e}")

# Plotando o modelo ajustado
plt.plot(V_fit, I, '--', label='Ajuste do modelo', color='blue', linewidth=2)

# Estilo científico
plt.xlabel('Tensão (V)', fontsize=24)
plt.ylabel('Corrente (A)', fontsize=24)
plt.title('Curva I-V com Margem de Erro\n(Modelo vs Experimental)', fontsize=24)
plt.legend(fontsize=18)
plt.grid(True, linestyle='--', alpha=0.7)

# Configurações adicionais
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tight_layout()
plt.show()
```
