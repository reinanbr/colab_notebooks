```python
date = table[1]['Data']
result = table[1]['Resultado'].fillna('NaN')
list(result)
```

```python
date = table[1]['Data']
result = list(table[1]['Posse'].fillna('NaN'))
result[:] = (term for term in result if term!='NaN')
(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats

plt.style.use('seaborn')

avarage = [stats.median(result) for _ in result]


# up avarage
up_ava = [i for i in result if i> avarage[0]]

#down avarage
down_ava = [i for i in result if i < avarage[0]]

print(f'up: {(100*(len(up_ava)/len(avarage))):.2f}%')
print(f'down: {(100*(len(down_ava)/len(avarage))):.2f}%')



# up 50%
up_ava = [i for i in result if i> 50]

#down 50%
down_ava = [i for i in result if i < 50]

print(f'up 50: {(100*(len(up_ava)/len(avarage))):.2f}%')
print(f'down 50: {(100*(len(down_ava)/len(avarage))):.2f}%')


fig, ax = plt.subplots()
ax.plot_date(date[:len(result)],avarage,'--',label=f'média da posse de bola nestes jogos\nmédia da posse de bola={avarage[0]}%\n')
ax.plot_date(date[:len(result)],result,'--o',label=f'% da posse de bola em cada partida\n{len(up_ava)} jogos com posse acima dos 50% \n[{(100*(len(up_ava)/len(avarage))):.2f}% dos jogos]\n{len(down_ava)} jogos com posse abaixo dos 50% \n[{(100*(len(down_ava)/len(avarage))):.2f}% dos jogos]')

plt.title(f'histórico da porcentagem da posse de bola do Corinthians\nnos últimos {len(result)} jogos',fontweight='bold')
plt.xlabel('data',fontweight='bold')
plt.ylabel('procentagem da posse de bola %',fontweight='bold')
plt.legend()
fig.autofmt_xdate()
plt.show()
#print(dir(date[1]))
#plt.xticks(date[:len(result)][::2],[f'{dat.day}/{dat.month}' for dat in date[:len(result)][::2]])
```

```python
from collections import Counter

count = Counter(list(result))

count.pop('NaN')
print(count)
def prob_laplace(count:Counter):
  values_sum = sum(list(count.values()))
  next_value = values_sum + 1
  d_value = count['D']/next_value
  e_value = count['E']/next_value
  v_value = count['V']/next_value
  error = 1-values_sum/next_value

  return {'D':d_value,'V':v_value,'E':e_value,'error':error}
```

```python
prob_laplace(count)
```
