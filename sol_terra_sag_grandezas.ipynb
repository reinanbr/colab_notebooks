```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import c, G

plt.style.use('classic')
plt.style.use('dark_background')
```

```python

def raio_sch(m):
  return 2*m*G/(c**2)


def vol(r):
  return (4/3)*np.pi*(r**3)

```

```python
raio_sag = 12_000_000_000
raio_sol = 696_340_000
raio_terra = 6_371_000

massa_sol = 1.989 * 10e30
massa_sag = 4_000_000 * massa_sol
massa_terra = 5.972 * 10e24

N = 1000

raio_list = np.linspace(raio_sol,raio_sag,N)
raio_sun = raio_list/raio_sol

mass_list = np.linspace(massa_sol,massa_sag,N)
mass_sun = mass_list/massa_sol

density_sun = (massa_sol/vol(raio_sol))
density_earth = (massa_terra/vol(raio_terra))/density_sun
density_sag = (massa_sag/vol(raio_sag))/density_sun


density = np.linspace(0,800,N)
print(density_sag,density_earth)
print(density_sun)
print(massa_sag/vol(raio_sag))
print( (massa_terra/vol(raio_terra)))
print(massa_sag)
```

```python


fig = plt.figure()
ax1 = plt.subplot(111)
ax2 = ax1.twinx()


ax1.plot(density,mass_sun,label='linha de densidade')
ax2.plot(density,raio_sun)


plt.annotate('Linha de Dens. de Sagitarius-A\nMa=7,9 x 10e37 kg\nRa = 12 10⁹ m\nDa = 10,9 x 10⁹ kg/m³\n781 x Ds',
             xy=(781,17),xytext=(450,7),
             size=11,
             arrowprops = dict(facecolor ='red',
                               lw=1.2,
                                  shrink = 0.001))



plt.annotate('Linha de Dens. do Sol \n 1 x Ms \n 1 x Rs \n 1 x Ds',
             xy=(1,1),xytext=(100,8),
             size=11,
             arrowprops = dict(facecolor ='darkorange',
                               lw=1.2,
                                  shrink = 0.001))


plt.annotate('Linha de Dens. da Terra\nMt = 5,9 x 10e24 kg\nRt = 6,3 x 10⁶ m\nDt = 55,1 x 10⁵ kg/m³  \n3,9 x Ds',
             xy=(3,0),xytext=(100,14),
             size=11,
             arrowprops = dict(facecolor ='green',
                               lw=1.2,
                                  shrink = 0.001))





ax1.axvline(density_sag,0,1,color='red')
ax1.set_ylabel('x Raio Solar  Rs = 6,96 x 10⁹ metros')

ax2.set_ylabel(r'x Massa Solar  Ms = 1,98 x $10^{30}$ kg')
ax2.axvline(x=density_earth,color='green')
ax2.axvline(x=1,color='darkorange')

ax1.set_xlabel('x Densidade Solar\nDs = 1.406 kg/m³')

ax1.grid()
ax2.grid()

ax1.legend(fontsize='10')

plt.title('relação de densidade\nTerra x Sol x Sagitarius A*',fontsize='12',fontweight='bold')
plt.savefig('rela.png',dpi=900)
```
