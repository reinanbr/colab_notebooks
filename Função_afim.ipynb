```python
a = 5
b = 7
import numpy as np
from sympy.plotting import*
import matplotlib.pyplot as plt
from sympy import*

x = Symbol("x")
y = Function("y")

y = a*x + b

#x = (-b)/a

# Não se converte expressão para numeros em programação
# expressão é Latex, serve apenas para print
# numeros são digitos decimais, estes sim são para fazer plotagens


y
```

```python

#em codigo, eh bem mais produtivo declarar as variaveis dentro do codigo

a = 5
b = 7

# funções para plotagem, devem seguir este tipo de nipe
# ou algo parecido
def afim(x): return a*x + b


t =np.linspace(0,10,100)

reta = afim(t) # criando o y para plotagem

plt.plot(reta) # do jeito que você fez
```

```python
#personalisando
plt.style.use('classic')
plt.style.use('seaborn')

plt.title('exemplo de gráfico de função afim',fontweight='bold')
plt.ylabel('y')
plt.xlabel('x')
plt.annotate(r'$y=5x+7$',size=25,xy=(2,40))

plt.plot(t,reta)
plt.show()
```
