```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("classic")
plt.style.use("seaborn-v0_8-darkgrid")

"""
UM EXEMPLO MUITO BOM É COM DADOS CLIMÁTICOS
"""


# Dados de exemplo
x = np.linspace(0, 10, 100)
temperatura = np.sin(x) * 10 + 20  # Exemplo de temperatura em Celsius
umidade = np.cos(x) * 20 + 50  # Exemplo de umidade relativa (%)
pressao = np.sin(x) * 5 + 1013  # Exemplo de pressão atmosférica (hPa)

fig, ax1 = plt.subplots()

# Primeiro eixo Y (esquerda)
ax1.set_xlabel("Tempo (dias)")
ax1.set_ylabel("Temperatura (°C)", color="tab:red")
ax1.plot(x, temperatura, color="tab:red", label="Temperatura")
ax1.tick_params(axis="y", labelcolor="tab:red")

# Segundo eixo Y (direita)
ax2 = ax1.twinx()
ax2.set_ylabel("Umidade (%)", color="tab:blue")
ax2.plot(x, umidade, color="tab:blue", linestyle="dashed", label="Umidade")
ax2.tick_params(axis="y", labelcolor="tab:blue")

# Terceira informação no mesmo eixo X (segunda linha no eixo esquerdo)
ax1.plot(x, pressao, color="tab:green", linestyle="dotted", label="Pressão")

# Adicionando legendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.grid(lw=".1")
plt.title("Variação de Temperatura, Umidade e Pressão ao Longo do Tempo")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8-darkgrid")

"""
VAMOS AUMENTANDO A SUTILEZA DO TRABALHO
"""

# Dados fictícios
dias = np.arange(1, 31)
preco = np.cumsum(np.random.randn(30) * 2) + 100  # Simulação de preço de ação
volume = np.random.randint(1000, 5000, size=30)  # Volume negociado
variacao = np.random.randn(30)  # Variação percentual diária

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo (preço e variação)
ax1.set_xlabel("Dias")
ax1.set_ylabel("Preço da Ação (USD)", color="tab:blue")
ax1.plot(dias, preco, color="tab:blue", marker="o", label="Preço da Ação")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Segundo eixo Y para volume
ax2 = ax1.twinx()
ax2.set_ylabel("Volume de Negociação", color="tab:green")
ax2.bar(dias, volume, color="tab:green", alpha=0.3, label="Volume")
ax2.tick_params(axis="y", labelcolor="tab:green")

# Adicionando variação percentual no eixo esquerdo
ax1.plot(dias, variacao * 10 + 100, color="tab:red", linestyle="dashed", label="Variação (%)")

# Legendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.title("Análise de Ações: Preço, Volume e Variação Percentual")
plt.grid(lw=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d

plt.style.use("seaborn-v0_8-darkgrid")

"""
REFINANDO A ANÁLISE PARA UMA MELHOR VISUALIZAÇÃO
"""

# Dados fictícios
dias = np.arange(1, 31)
preco_bruto = np.cumsum(np.random.randn(30) * 2) + 100  # Simulação de preço de ação
preco = gaussian_filter1d(preco_bruto, sigma=2)  # Aplicando suavização
volume = np.random.randint(1000, 5000, size=30)  # Volume negociado
variacao = np.random.randn(30)  # Variação percentual diária

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo (preço e variação)
ax1.set_xlabel("Dias", fontsize=12)
ax1.set_ylabel("Preço da Ação (USD)", color="tab:blue", fontsize=12)
ax1.plot(dias, preco, color="tab:blue", marker="o", linestyle="-", label="Preço da Ação", linewidth=2)
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Segundo eixo Y para volume
ax2 = ax1.twinx()
ax2.set_ylabel("Volume de Negociação", color="tab:gray", fontsize=12)
ax2.bar(dias, volume, color="gray", alpha=0.3, label="Volume")
ax2.tick_params(axis="y", labelcolor="gray")

# Adicionando variação percentual no eixo esquerdo
ax1.plot(dias, gaussian_filter1d(variacao * 10 + 100, sigma=1.5),
         color="tab:red", linestyle="dashed", linewidth=2, label="Variação (%)")

# Adicionando rótulos sutis no gráfico
for i in range(0, len(dias), 5):
    ax1.text(dias[i], preco[i] + 1, f"{preco[i]:.1f}", color="tab:blue", fontsize=8)
    ax2.text(dias[i], volume[i] + 200, f"{volume[i]}", color="gray", fontsize=8)

# Legendas
ax1.legend(loc="upper left", fontsize=10)
ax2.legend(loc="upper right", fontsize=10)

# Melhorando layout e aparência
plt.title("Análise Refinada: Preço, Volume e Variação Percentual", fontsize=14, fontweight="bold")
plt.grid(lw=0.2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")


# Definição dos parâmetros
x = np.linspace(-10, 10, 200)  # Espaço unidimensional
t = np.linspace(0, 5, 100)  # Tempo
X, T = np.meshgrid(x, t)

# Campo escalar φ(x, t) representado como uma onda oscilante
phi = np.sin(X - T) * np.exp(-0.1 * (X - T)**2)

# Energia associada ao campo (proporcional a φ²)
energia = phi**2

# Densidade de probabilidade (normalizada)
densidade_probabilidade = np.abs(phi) / np.max(np.abs(phi))

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo: Campo escalar φ(x,t)
ax1.set_xlabel("Posição (x)", fontsize=12)
ax1.set_ylabel("Campo Escalar ϕ(x,t)", color="tab:blue", fontsize=12)
ax1.plot(x, phi[50, :], color="tab:blue", linestyle="-", linewidth=2, label="Campo ϕ(x,t)")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Eixo direito: Energia do campo
ax2 = ax1.twinx()
ax2.set_ylabel("Energia do Campo", color="tab:red", fontsize=12)
ax2.plot(x, energia[50, :], color="tab:red", linestyle="dashed", linewidth=2, label="Energia ϕ²")
ax2.tick_params(axis="y", labelcolor="tab:red")

# Adicionando densidade de probabilidade como uma terceira informação
ax1.fill_between(x, densidade_probabilidade[50, :] * np.max(phi), alpha=0.2, color="purple", label="Densidade de Probabilidade")

# Legendas
ax1.legend(loc="upper left", fontsize=10)
ax2.legend(loc="upper right", fontsize=10)

# Ajustes finais
plt.title("Evolução de um Campo Escalar na Teoria Quântica de Campos", fontsize=14, fontweight="bold")
plt.grid(lw=0.2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")


# Constantes de normalização (valores arbitrários para ilustração)
rho_rad_0 = 1e15  # Densidade inicial de radiação
rho_mat_0 = 1e12  # Densidade inicial de matéria
f_b = 0.16  # Fração bariônica da matéria (aproximadamente 16%)

# Tempo após o Big Bang em segundos (escala logarítmica)
tempo = np.logspace(-12, 4, 100)  # De 10⁻¹² s a 10⁴ s

# Densidade de radiação (dominante no início)
rho_rad = rho_rad_0 * tempo**-2

# Densidade de matéria (dominante posteriormente)
rho_mat = rho_mat_0 * tempo**-1.5

# Densidade total é a soma das densidades de radiação e matéria
rho_total = rho_rad + rho_mat

# Densidade bariônica é uma fração da densidade de matéria
rho_b = f_b * rho_mat

# Fração da densidade bariônica sobre a densidade total
frac_b = rho_b / rho_total

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo: Densidades
ax1.set_xlabel("Tempo após o Big Bang (s) [Escala Log]", fontsize=12)
ax1.set_ylabel("Densidade (unidades arbitrárias)", fontsize=12)
ax1.loglog(tempo, rho_total, color="tab:blue", linestyle="-", linewidth=2, label="Densidade Total ρ_total")
ax1.loglog(tempo, rho_b, color="tab:red", linestyle="dashed", linewidth=2, label="Densidade Bariônica ρ_b")
ax1.tick_params(axis="y")

# Eixo direito: Fração ρ_b / ρ_total
ax2 = ax1.twinx()
ax2.set_ylabel("Fração ρ_b / ρ_total", color="purple", fontsize=12)
ax2.semilogx(tempo, frac_b, color="purple", linestyle="dotted", linewidth=2, label="Fração Bariônica ρ_b / ρ_total")
ax2.tick_params(axis="y", labelcolor="purple")

# Legendas
ax1.legend(loc="upper right", fontsize=10)
ax2.legend(loc="lower left", fontsize=10)

# Ajustes visuais
plt.title("Evolução da Densidade Bariônica e Total Após o Big Bang", fontsize=14, fontweight="bold")
plt.grid(lw=0.2, which="both")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")

tempo = np.logspace(-12, 17, 200)

rho_rad_0 = 1e15
rho_mat_0 = 1e12
rho_lambda_0 = 1e-30
f_b = 0.16

rho_rad = rho_rad_0 * tempo**-2
rho_mat = rho_mat_0 * tempo**-1.5
rho_lambda = np.full_like(tempo, rho_lambda_0)
rho_total = rho_rad + rho_mat + rho_lambda
rho_b = f_b * rho_mat
frac_b = rho_b / rho_total

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_xlabel("Tempo (s) [Escala Log]", fontsize=12)
ax1.set_ylabel("Densidade (unidades arbitrárias)", fontsize=12)
ax1.loglog(tempo, rho_total, color="tab:blue", linestyle="-", linewidth=2, label="Densidade Total ρ_total")
ax1.loglog(tempo, rho_rad, color="tab:orange", linestyle="dashed", linewidth=2, label="Densidade de Radiação ρ_rad")
ax1.loglog(tempo, rho_mat, color="tab:green", linestyle="dashdot", linewidth=2, label="Densidade de Matéria ρ_mat")
ax1.loglog(tempo, rho_lambda, color="tab:brown", linestyle="dotted", linewidth=2, label="Densidade da Energia Escura ρ_Λ")
ax1.loglog(tempo, rho_b, color="tab:red", linestyle="dashed", linewidth=2, label="Densidade Bariônica ρ_b")
ax1.tick_params(axis="y")

ax2 = ax1.twinx()
ax2.set_ylabel("Fração ρ_b / ρ_total", color="purple", fontsize=12)
ax2.semilogx(tempo, frac_b, color="purple", linestyle="dotted", linewidth=2, label="Fração Bariônica ρ_b / ρ_total")
ax2.tick_params(axis="y", labelcolor="purple")

ax1.legend(loc="upper right", fontsize=10)
ax2.legend(loc="lower left", fontsize=10)

plt.title("Evolução das Densidades Cosmológicas do Big Bang aos Dias Atuais\n(Registro Do Momento Em Que Deus Criou O Universo Em Que Vivemos)", fontsize=14, fontweight="bold")
plt.grid(lw=0.2, which="both")
plt.tight_layout()
plt.show()
```
