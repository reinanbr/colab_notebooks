```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("classic")
plt.style.use("seaborn-v0_8-darkgrid")

"""
UM EXEMPLO MUITO BOM  COM DADOS CLIMTICOS
"""


# Dados de exemplo
x = np.linspace(0, 10, 100)
temperatura = np.sin(x) * 10 + 20  # Exemplo de temperatura em Celsius
umidade = np.cos(x) * 20 + 50  # Exemplo de umidade relativa (%)
pressao = np.sin(x) * 5 + 1013  # Exemplo de presso atmosfrica (hPa)

fig, ax1 = plt.subplots()

# Primeiro eixo Y (esquerda)
ax1.set_xlabel("Tempo (dias)")
ax1.set_ylabel("Temperatura (C)", color="tab:red")
ax1.plot(x, temperatura, color="tab:red", label="Temperatura")
ax1.tick_params(axis="y", labelcolor="tab:red")

# Segundo eixo Y (direita)
ax2 = ax1.twinx()
ax2.set_ylabel("Umidade (%)", color="tab:blue")
ax2.plot(x, umidade, color="tab:blue", linestyle="dashed", label="Umidade")
ax2.tick_params(axis="y", labelcolor="tab:blue")

# Terceira informao no mesmo eixo X (segunda linha no eixo esquerdo)
ax1.plot(x, pressao, color="tab:green", linestyle="dotted", label="Presso")

# Adicionando legendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.grid(lw=".1")
plt.title("Variao de Temperatura, Umidade e Presso ao Longo do Tempo")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8-darkgrid")

"""
VAMOS AUMENTANDO A SUTILEZA DO TRABALHO
"""

# Dados fictcios
dias = np.arange(1, 31)
preco = np.cumsum(np.random.randn(30) * 2) + 100  # Simulao de preo de ao
volume = np.random.randint(1000, 5000, size=30)  # Volume negociado
variacao = np.random.randn(30)  # Variao percentual diria

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo (preo e variao)
ax1.set_xlabel("Dias")
ax1.set_ylabel("Preo da Ao (USD)", color="tab:blue")
ax1.plot(dias, preco, color="tab:blue", marker="o", label="Preo da Ao")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Segundo eixo Y para volume
ax2 = ax1.twinx()
ax2.set_ylabel("Volume de Negociao", color="tab:green")
ax2.bar(dias, volume, color="tab:green", alpha=0.3, label="Volume")
ax2.tick_params(axis="y", labelcolor="tab:green")

# Adicionando variao percentual no eixo esquerdo
ax1.plot(dias, variacao * 10 + 100, color="tab:red", linestyle="dashed", label="Variao (%)")

# Legendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.title("Anlise de Aes: Preo, Volume e Variao Percentual")
plt.grid(lw=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d

plt.style.use("seaborn-v0_8-darkgrid")

"""
REFINANDO A ANLISE PARA UMA MELHOR VISUALIZAO
"""

# Dados fictcios
dias = np.arange(1, 31)
preco_bruto = np.cumsum(np.random.randn(30) * 2) + 100  # Simulao de preo de ao
preco = gaussian_filter1d(preco_bruto, sigma=2)  # Aplicando suavizao
volume = np.random.randint(1000, 5000, size=30)  # Volume negociado
variacao = np.random.randn(30)  # Variao percentual diria

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo (preo e variao)
ax1.set_xlabel("Dias", fontsize=12)
ax1.set_ylabel("Preo da Ao (USD)", color="tab:blue", fontsize=12)
ax1.plot(dias, preco, color="tab:blue", marker="o", linestyle="-", label="Preo da Ao", linewidth=2)
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Segundo eixo Y para volume
ax2 = ax1.twinx()
ax2.set_ylabel("Volume de Negociao", color="tab:gray", fontsize=12)
ax2.bar(dias, volume, color="gray", alpha=0.3, label="Volume")
ax2.tick_params(axis="y", labelcolor="gray")

# Adicionando variao percentual no eixo esquerdo
ax1.plot(dias, gaussian_filter1d(variacao * 10 + 100, sigma=1.5),
         color="tab:red", linestyle="dashed", linewidth=2, label="Variao (%)")

# Adicionando rtulos sutis no grfico
for i in range(0, len(dias), 5):
    ax1.text(dias[i], preco[i] + 1, f"{preco[i]:.1f}", color="tab:blue", fontsize=8)
    ax2.text(dias[i], volume[i] + 200, f"{volume[i]}", color="gray", fontsize=8)

# Legendas
ax1.legend(loc="upper left", fontsize=10)
ax2.legend(loc="upper right", fontsize=10)

# Melhorando layout e aparncia
plt.title("Anlise Refinada: Preo, Volume e Variao Percentual", fontsize=14, fontweight="bold")
plt.grid(lw=0.2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")


# Definio dos parmetros
x = np.linspace(-10, 10, 200)  # Espao unidimensional
t = np.linspace(0, 5, 100)  # Tempo
X, T = np.meshgrid(x, t)

# Campo escalar (x, t) representado como uma onda oscilante
phi = np.sin(X - T) * np.exp(-0.1 * (X - T)**2)

# Energia associada ao campo (proporcional a )
energia = phi**2

# Densidade de probabilidade (normalizada)
densidade_probabilidade = np.abs(phi) / np.max(np.abs(phi))

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo: Campo escalar (x,t)
ax1.set_xlabel("Posio (x)", fontsize=12)
ax1.set_ylabel("Campo Escalar (x,t)", color="tab:blue", fontsize=12)
ax1.plot(x, phi[50, :], color="tab:blue", linestyle="-", linewidth=2, label="Campo (x,t)")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# Eixo direito: Energia do campo
ax2 = ax1.twinx()
ax2.set_ylabel("Energia do Campo", color="tab:red", fontsize=12)
ax2.plot(x, energia[50, :], color="tab:red", linestyle="dashed", linewidth=2, label="Energia ")
ax2.tick_params(axis="y", labelcolor="tab:red")

# Adicionando densidade de probabilidade como uma terceira informao
ax1.fill_between(x, densidade_probabilidade[50, :] * np.max(phi), alpha=0.2, color="purple", label="Densidade de Probabilidade")

# Legendas
ax1.legend(loc="upper left", fontsize=10)
ax2.legend(loc="upper right", fontsize=10)

# Ajustes finais
plt.title("Evoluo de um Campo Escalar na Teoria Quntica de Campos", fontsize=14, fontweight="bold")
plt.grid(lw=0.2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")


# Constantes de normalizao (valores arbitrrios para ilustrao)
rho_rad_0 = 1e15  # Densidade inicial de radiao
rho_mat_0 = 1e12  # Densidade inicial de matria
f_b = 0.16  # Frao barinica da matria (aproximadamente 16%)

# Tempo aps o Big Bang em segundos (escala logartmica)
tempo = np.logspace(-12, 4, 100)  # De 10 s a 10 s

# Densidade de radiao (dominante no incio)
rho_rad = rho_rad_0 * tempo**-2

# Densidade de matria (dominante posteriormente)
rho_mat = rho_mat_0 * tempo**-1.5

# Densidade total  a soma das densidades de radiao e matria
rho_total = rho_rad + rho_mat

# Densidade barinica  uma frao da densidade de matria
rho_b = f_b * rho_mat

# Frao da densidade barinica sobre a densidade total
frac_b = rho_b / rho_total

fig, ax1 = plt.subplots(figsize=(10, 5))

# Eixo esquerdo: Densidades
ax1.set_xlabel("Tempo aps o Big Bang (s) [Escala Log]", fontsize=12)
ax1.set_ylabel("Densidade (unidades arbitrrias)", fontsize=12)
ax1.loglog(tempo, rho_total, color="tab:blue", linestyle="-", linewidth=2, label="Densidade Total _total")
ax1.loglog(tempo, rho_b, color="tab:red", linestyle="dashed", linewidth=2, label="Densidade Barinica _b")
ax1.tick_params(axis="y")

# Eixo direito: Frao _b / _total
ax2 = ax1.twinx()
ax2.set_ylabel("Frao _b / _total", color="purple", fontsize=12)
ax2.semilogx(tempo, frac_b, color="purple", linestyle="dotted", linewidth=2, label="Frao Barinica _b / _total")
ax2.tick_params(axis="y", labelcolor="purple")

# Legendas
ax1.legend(loc="upper right", fontsize=10)
ax2.legend(loc="lower left", fontsize=10)

# Ajustes visuais
plt.title("Evoluo da Densidade Barinica e Total Aps o Big Bang", fontsize=14, fontweight="bold")
plt.grid(lw=0.2, which="both")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")

tempo = np.logspace(-12, 17, 200)

rho_rad_0 = 1e15
rho_mat_0 = 1e12
rho_lambda_0 = 1e-30
f_b = 0.16

rho_rad = rho_rad_0 * tempo**-2
rho_mat = rho_mat_0 * tempo**-1.5
rho_lambda = np.full_like(tempo, rho_lambda_0)
rho_total = rho_rad + rho_mat + rho_lambda
rho_b = f_b * rho_mat
frac_b = rho_b / rho_total

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_xlabel("Tempo (s) [Escala Log]", fontsize=12)
ax1.set_ylabel("Densidade (unidades arbitrrias)", fontsize=12)
ax1.loglog(tempo, rho_total, color="tab:blue", linestyle="-", linewidth=2, label="Densidade Total _total")
ax1.loglog(tempo, rho_rad, color="tab:orange", linestyle="dashed", linewidth=2, label="Densidade de Radiao _rad")
ax1.loglog(tempo, rho_mat, color="tab:green", linestyle="dashdot", linewidth=2, label="Densidade de Matria _mat")
ax1.loglog(tempo, rho_lambda, color="tab:brown", linestyle="dotted", linewidth=2, label="Densidade da Energia Escura _")
ax1.loglog(tempo, rho_b, color="tab:red", linestyle="dashed", linewidth=2, label="Densidade Barinica _b")
ax1.tick_params(axis="y")

ax2 = ax1.twinx()
ax2.set_ylabel("Frao _b / _total", color="purple", fontsize=12)
ax2.semilogx(tempo, frac_b, color="purple", linestyle="dotted", linewidth=2, label="Frao Barinica _b / _total")
ax2.tick_params(axis="y", labelcolor="purple")

ax1.legend(loc="upper right", fontsize=10)
ax2.legend(loc="lower left", fontsize=10)

plt.title("Evoluo das Densidades Cosmolgicas do Big Bang aos Dias Atuais\n(Registro Do Momento Em Que Deus Criou O Universo Em Que Vivemos)", fontsize=14, fontweight="bold")
plt.grid(lw=0.2, which="both")
plt.tight_layout()
plt.show()
```
