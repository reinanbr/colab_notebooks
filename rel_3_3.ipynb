```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')

# --- Dados experimentais ---
time = np.array([0, 30, 60, 90, 120, 150, 180, 210, 240, 270,
                 300, 330, 360, 390, 420, 450, 480, 510, 570, 600])
Vc_exp = np.array([6.3, 5.19, 4.22, 3.43, 2.82, 2.32, 1.82, 1.42, 1.21, 0.97,
                   0.79, 0.64, 0.52, 0.40, 0.35, 0.29, 0.24, 0.20, 0.16, 0.14])
Vr_exp = np.array([-6.40, -5.27, -3.48, -4.32, -2.86, -2.35, -1.85, -1.44, -1.22, -0.98,
                   -0.8, -0.64, -0.52, -0.40, -0.35, -0.29, -0.24, -0.19, -0.16, -0.13])

# --- Parâmetros do circuito ---
V_max = 7.68      # tensão máxima do capacitor (V)
R = 147900        # ohms
C = 1.25e-3       # farads
tau = R * C       # constante de tempo

# --- Curvas teóricas ---
t_fit = np.linspace(0, 600, 500)
Vc_fit = V_max * np.exp(-t_fit / tau)
Vr_fit = -Vc_fit

# --- Plot ---
plt.figure(figsize=(5, 6), facecolor='black', dpi=600)
ax = plt.gca()
ax.set_facecolor('black')

plt.plot(time, Vc_exp, 'o', color='lime', label='Vc experimental')
plt.plot(t_fit, Vc_fit, '--', color='cyan', label='Vc teórico')

plt.plot(time, Vr_exp, 's', color='red', label='Vr experimental')
plt.plot(t_fit, Vr_fit, '--', color='orange', label='Vr teórico')

# --- Equações no gráfico ---
equacao_ = r'$V_C(t) = V_{\rm max} e^{-t/(RC)}$'
plt.text(300, max(Vc_exp)*0.98, equacao_, color='gold', fontsize=17, bbox=dict(facecolor='black', alpha=0.5))

equacao = f'$V_C(t) = {V_max:.2f} e^{{-t/({tau:.2f})}}$'
plt.text(180, max(Vc_exp)*0.5, equacao, color='cyan', fontsize=18, bbox=dict(facecolor='black', alpha=0.5))

# --- Labels e título ---
plt.xlabel(r'$Tempo\ (s)$', color='white', fontsize=14)
plt.ylabel(r'$Tensão\ (V)$', color='white', fontsize=14)
plt.title('Descarregamento do Capacitor\n'+r'$R = 147.9K\Omega$'+' | '+r'$V_{\rm max} = 7.68V$', color='white')

plt.grid(True, linestyle='--', alpha=0.3, color='white')
plt.tick_params(colors='white')

# --- Legenda ---
legend = ax.legend(loc='lower right')
legend.get_frame().set_facecolor('black')
legend.get_frame().set_edgecolor('gold')
for txt, col in zip(legend.get_texts(), ['lime','cyan','red','orange']):
    txt.set_color(col)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import stdev, mean
plt.style.use('classic')

# --- Dados ---
Vr = np.array([-6.40, -5.27, -3.48, -4.32, -2.86, -2.35, -1.85, -1.44, -1.22, -0.98,
               -0.8, -0.64, -0.52, -0.40, -0.35, -0.29, -0.24, -0.19, -0.16, -0.13])
Vc = np.array([6.3, 5.19, 4.22, 3.43, 2.82, 2.32, 1.82, 1.42, 1.21, 0.97,
               0.79, 0.64, 0.52, 0.40, 0.35, 0.29, 0.24, 0.20, 0.16, 0.14])

# --- Soma ---
rc_ = Vr + Vc

# --- Estatística ---
m_rc = mean(rc_)
desv = stdev(rc_)
m_rc_minus_ = np.ones(len(rc_)+2)*(m_rc-desv)
m_rc_plus_ = np.ones(len(rc_)+2)*(m_rc+desv)

time = np.linspace(0, 600, len(Vr))
time_ = np.linspace(-30, 630, len(Vr)+2)

print(rc_)
print('média:', m_rc, '\t', 'desvio:', desv)

# --- Gráfico ---
plt.figure(figsize=(5, 6), facecolor='black',dpi=600)
ax = plt.gca()
ax.set_facecolor('black')

plt.title("Soma das Tensões no Resistor e no Capacitor", color='white', fontsize=12, pad=15)
plt.ylabel('Tensão (V)', color='white')
plt.xlabel('Tempo (s)', color='white')
plt.grid(True, linestyle='--', alpha=0.4, color='white')
plt.tick_params(colors='white')

# --- Plotagem ---
plt.plot(time, rc_, 'o', color='cyan', label='Soma experimental V_R + V_C')
plt.plot(time_, m_rc*np.ones(len(time_)), 'o-', c='green', label=f'média: {m_rc:.2f}')
plt.plot(time_, m_rc_minus_, '--', c='yellow', label=f'desvio padrão: {m_rc:.2f} $\mp$ {desv:.2f}')
plt.plot(time_, m_rc_plus_, '--', c='yellow')

# --- Legenda ---
legend = plt.legend(facecolor='black', edgecolor='white', fontsize=9)
for text in legend.get_texts():
    text.set_color("white")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev
plt.style.use('default')

# --- Dados ---
Vr = np.array([-6.40, -5.27, -3.48, -4.32, -2.86, -2.35, -1.85, -1.44, -1.22, -0.98,
               -0.8, -0.64, -0.52, -0.40, -0.35, -0.29, -0.24, -0.19, -0.16, -0.13])
Vc = np.array([6.3, 5.19, 4.22, 3.43, 2.82, 2.32, 1.82, 1.42, 1.21, 0.97,
               0.79, 0.64, 0.52, 0.40, 0.35, 0.29, 0.24, 0.20, 0.16, 0.14])

# --- Soma ---
rc_ = Vr + Vc

# --- Estatística ---
m_rc = mean(rc_)
desv = stdev(rc_)
m_rc_minus_ = np.ones(len(rc_)+2)*(m_rc-desv)
m_rc_plus_ = np.ones(len(rc_)+2)*(m_rc+desv)

# --- Tempo ---
time = np.linspace(0, 600, len(rc_))
time_ = np.linspace(-30, 630, len(rc_)+2)

print(rc_)
print('média:', m_rc, '\t', 'desvio:', desv)

# --- Plotagem ---
plt.figure(figsize=(6,6), dpi=600)
plt.errorbar(time, rc_, yerr=np.abs(rc_-m_rc), fmt='o', ecolor='purple', capsize=5, label='Soma V_R + V_C')
plt.axhline(m_rc, color='red', linestyle='--', label=f'Média ({m_rc:.2f})')
plt.plot(time_, m_rc_minus_, '--', c='orange', label=f'Desvio padrão ±{desv:.2f}')
plt.plot(time_, m_rc_plus_, '--', c='orange')

# --- Anotar discrepância em cada ponto ---
for t, val in zip(time, rc_):
    plt.text(t, val + 0.05, f'{val-m_rc:+.2f}', ha='center', va='bottom', fontsize=8, color='blue')

plt.xlabel('Tempo (s)')
plt.ylabel('Tensão V_R + V_C (V)')
plt.title('Soma das Tensões no Resistor e no Capacitor\ncom Barras de Erro e Discrepância')
plt.legend()
plt.grid(True)
plt.show()
```
